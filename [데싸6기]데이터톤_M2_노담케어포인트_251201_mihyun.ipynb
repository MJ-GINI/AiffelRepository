# í”¼ì²˜ ìƒì„±

```python
# ë‚˜ì´ êµ¬ê°„í™” (Binning)
# ë‚˜ì´ì— ë”°ë¥¸ ë¹„ì„ í˜•ì ì¸ ê´€ê³„ë¥¼ ëª¨ë¸ì´ í•™ìŠµ

bins = [10*i for i in range(2, 10)]
labels = [f'{i}-{i+9}' for i in bins[:-1]]

train['age_group'] = pd.cut(train['age'], bins=bins, labels=labels, right=False)
test['age_group'] = pd.cut(test['age'], bins=bins, labels=labels, right=False)

print("Train DataFrame 'age_group' head:")
print(train['age_group'].head())
print("\nTest DataFrame 'age_group' head:")
print(test['age_group'].head())
```

```python
# ì²´í˜• ê´€ë ¨ íŒŒìƒ ë³€ìˆ˜ ìƒì„±

train['BMI'] = train['weight(kg)'] / ((train['height(cm)'] / 100) ** 2)
test['BMI'] = test['weight(kg)'] / ((test['height(cm)'] / 100) ** 2)

train['waist_to_height_ratio'] = train['waist(cm)'] / train['height(cm)'] # ì¶”ê°€
test['waist_to_height_ratio'] = test['waist(cm)'] / test['height(cm)']

train['weight_waist_interaction'] = train['weight(kg)'] * train['waist(cm)']
test['weight_waist_interaction'] = test['weight(kg)'] * test['waist(cm)']

train['BMI_AST_interaction'] = train['BMI'] * train['AST']
test['BMI_AST_interaction'] = test['BMI'] * test['AST']

train['BMI_ALT_interaction'] = train['BMI'] * train['ALT']
test['BMI_ALT_interaction'] = test['BMI'] * test['ALT']

train['BMI_Gtp_interaction'] = train['BMI'] * train['Gtp']
test['BMI_Gtp_interaction'] = test['BMI'] * test['Gtp']
```

```python
# ì½œë ˆìŠ¤í…Œë¡¤ ê´€ë ¨ íŒŒìƒ ë³€ìˆ˜ ìƒì„±

train['HDL_risk'] = train['HDL'] / train['LDL'] # ë†’ì„ìˆ˜ë¡ ì¢‹ë‹¤(???)
test['HDL_risk'] = test['HDL'] / test['LDL']

train['TG_risk'] = train['triglyceride'] / train['HDL'] # ë‚®ì„ìˆ˜ë¡ ì¢‹ë‹¤(???)
test['TG_risk'] = test['triglyceride'] / test['HDL']
```

```python
# ì‹œë ¥ ë° ì²­ë ¥ ê´€ë ¨ íŒŒ ë³€ìˆ˜ ìƒì„±

train['avg_eyesight'] = (train['eyesight(left)'] + train['eyesight(right)']) / 2
test['avg_eyesight'] = (test['eyesight(left)'] + test['eyesight(right)']) / 2

train['impaired_eyesight'] = (train['avg_eyesight'] < 0.5).astype(int)
test['impaired_eyesight'] = (test['avg_eyesight'] < 0.5).astype(int)

train['impaired_hearing'] = ((train['hearing(left)'] != 1) | (train['hearing(right)'] != 1)).astype(int)
test['impaired_hearing'] = ((test['hearing(left)'] != 1) | (test['hearing(right)'] != 1)).astype(int)
```

```python
# í˜ˆì•• ê´€ë ¨ íŒŒìƒ ë³€ìˆ˜ ìƒì„±

train['pulse_pressure'] = train['systolic'] - train['relaxation']
test['pulse_pressure'] = test['systolic'] - test['relaxation']
```

```python
# ë¡œê·¸ ë³€í™˜ 'Gtp'

import numpy as np

# 'Gtp' ì¹¼ëŸ¼ì— ë¡œê·¸ ë³€í™˜ ì ìš©
train['Gtp'] = np.log1p(train['Gtp'])
test['Gtp'] = np.log1p(test['Gtp'])

print("Train DataFrame 'Gtp' after log transformation (head):")
display(train['Gtp'].head())
print("\nTest DataFrame 'Gtp' after log transformation (head):")
display(test['Gtp'].head())
```

# ë°ì´í„° ìŠ¤ì¼€ì¼ë§ ë° ì¸ì½”ë”©

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

# íƒ€ê²Ÿ ë³€ìˆ˜ ë¶„ë¦¬
X_train = train.drop('smoking', axis=1)
y_train = train['smoking']
X_test = test.copy()

# ìˆ«ìží˜• ë° ë²”ì£¼í˜• í”¼ì²˜ ì •ì˜
numerical_features = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = ['age_group']

# ColumnTransformer ì„¤ì •
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# ì „ì²˜ë¦¬ íŒŒì´í”„ë¼ì¸ ìƒì„± ë° ì ìš©
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# OneHotEncoderë¡œ ìƒì„±ëœ ìƒˆë¡œìš´ í”¼ì²˜ ì´ë¦„ ê°€ì ¸ì˜¤ê¸°
encoded_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)

# ëª¨ë“  í”¼ì²˜ ì´ë¦„ ì¡°í•©
all_feature_names = numerical_features + encoded_feature_names.tolist()

# ì²˜ë¦¬ëœ ë°ì´í„°ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜ (ì„ íƒ ì‚¬í•­, ëª¨ë¸ì— ë”°ë¼ numpy ë°°ì—´ë¡œ ì‚¬ìš© ê°€ëŠ¥)
X_train_df = pd.DataFrame(X_train_processed, columns=all_feature_names)
X_test_df = pd.DataFrame(X_test_processed, columns=all_feature_names)

print("X_train_df head after scaling and encoding:")
display(X_train_df.head())
print("\nX_test_df head after scaling and encoding:")
display(X_test_df.head())
```

# ëª¨ë¸ë§

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

# lr
log_reg_model = LogisticRegression(solver='liblinear', random_state=42, n_jobs=-1)

# rf
rf_model = RandomForestClassifier(random_state=42, n_jobs=-1)

# Igbm
lgbm_model = LGBMClassifier(random_state=42, n_jobs=-1)

# xgb
xgb_model = XGBClassifier(random_state=42, n_jobs=-1, eval_metric='logloss')

# Train lr
print("Training Logistic Regression model...")
log_reg_model.fit(X_train_processed, y_train)
print("Logistic Regression model trained.")

# Train rf
print("Training RandomForestClassifier model...")
rf_model.fit(X_train_processed, y_train)
print("RandomForestClassifier model trained.")

# Train lgbm
print("Training LGBMClassifier model...")
lgbm_model.fit(X_train_processed, y_train)
print("LGBMClassifier model trained.")

# Train xgb
print("Training XGBClassifier model...")
xgb_model.fit(X_train_processed, y_train)
print("XGBClassifier model trained.")
```

```python
from sklearn.metrics import roc_auc_score

# Evaluate lr model
y_pred_log_reg = log_reg_model.predict_proba(X_train_processed)[:, 1]
auc_log_reg = roc_auc_score(y_train, y_pred_log_reg)
print(f"Logistic Regression AUC-ROC: {auc_log_reg:.4f}")

# Evaluate rf model
y_pred_rf = rf_model.predict_proba(X_train_processed)[:, 1]
auc_rf = roc_auc_score(y_train, y_pred_rf)
print(f"RandomForestClassifier AUC-ROC: {auc_rf:.4f}")

# Evaluate lgbm model
y_pred_lgbm = lgbm_model.predict_proba(X_train_processed)[:, 1]
auc_lgbm = roc_auc_score(y_train, y_pred_lgbm)
print(f"LGBMClassifier AUC-ROC: {auc_lgbm:.4f}")

# Evaluate xgb model
y_pred_xgb = xgb_model.predict_proba(X_train_processed)[:, 1]
auc_xgb = roc_auc_score(y_train, y_pred_xgb)
print(f"XGBClassifier AUC-ROC: {auc_xgb:.4f}")
```

### ëžœë¤ê·¸ë¦¬ë“œì„œì¹˜

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint

# XGBoost ëª¨ë¸ì˜ í•˜ì´í¼íŒŒë¼ë¯¸í„° íƒìƒ‰ ê³µê°„ ì •ì˜
# ì´ì „ í›ˆë ¨ì—ì„œ ì‚¬ìš©í•œ XGBClassifier ì¸ìŠ¤í„´ìŠ¤ë¥¼ ì§ì ‘ ì‚¬ìš©

param_dist = {
    'n_estimators': randint(100, 500),  # ë¶€ìŠ¤íŒ… ë¼ìš´ë“œ ìˆ˜
    'learning_rate': uniform(0.01, 0.2), # í•™ìŠµë¥ 
    'max_depth': randint(3, 10),        # íŠ¸ë¦¬ì˜ ìµœëŒ€ ê¹Šì´
    'subsample': uniform(0.6, 0.4),     # ìƒ˜í”Œ ë¹„ìœ¨
    'colsample_bytree': uniform(0.6, 0.4), # í”¼ì²˜ ìƒ˜í”Œ ë¹„ìœ¨
    'gamma': uniform(0, 0.5),           # ë¦¬í”„ ë…¸ë“œ ë¶„í• ì— í•„ìš”í•œ ìµœì†Œ ì†ì‹¤ ê°ì†Œ
    'min_child_weight': randint(1, 10) # ìžì‹ ë…¸ë“œì— í•„ìš”í•œ ìµœì†Œ ê°€ì¤‘ì¹˜ í•©
}

# RandomizedSearchCV ì„¤ì •
# cv=3ì€ ë¹ ë¥´ê²Œ íƒìƒ‰í•˜ê¸° ìœ„í•¨ (ì‹¤ì œë¡œëŠ” 5 ì´ìƒ ê¶Œìž¥)
# n_iter=50ì€ íƒìƒ‰í•  í•˜ì´í¼íŒŒë¼ë¯¸í„° ì¡°í•©ì˜ ìˆ˜ (ë¹ ë¥¸ í™•ì¸ì„ ìœ„í•¨)
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=50,
    scoring='roc_auc',
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=-1 # ëª¨ë“  ì½”ì–´ ì‚¬ìš©
)

print("Starting RandomizedSearchCV for XGBoost model...")
# íŠœë‹ ì‹¤í–‰
random_search.fit(X_train_processed, y_train)

print("RandomizedSearchCV completed.")
print(f"Best AUC-ROC score: {random_search.best_score_:.4f}")
print("Best parameters found:")
display(random_search.best_params_)
```

### ê·¸ë¦¬ë“œì„œì¹˜

```python
from sklearn.model_selection import GridSearchCV

# Define a narrower parameter grid around the best parameters found by RandomizedSearchCV
param_grid = {
    'n_estimators': [350, 375, 400],
    'learning_rate': [0.03, 0.04, 0.05],
    'max_depth': [7, 8, 9],
    'subsample': [0.75, 0.80, 0.85],
    'colsample_bytree': [0.70, 0.75, 0.80],
    'gamma': [0.3, 0.4, 0.5],
    'min_child_weight': [1, 2]
}

print("GridSearchCV parameter grid defined:")
print(param_grid)
```

## ìµœì  íŒŒë¼ë¯¸í„°ë¡œ XGBoost ëª¨ë¸ ìž¬í•™ìŠµ ë° í‰ê°€ - ë¯¸í˜„ ìˆ˜ì • ë²„ì „

```python
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from scipy.stats import uniform, randint
import xgboost as xgb
import time

# X_train_processed, y_train, xgb_model (ë² ì´ìŠ¤ë¼ì¸ ëª¨ë¸)ì€ ì´ë¯¸ ì •ì˜ë˜ì–´ ìžˆë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤.

print("--- ðŸ§  ê³¼ì í•© ë°©ì§€ XGBoost íŠœë‹ (RandomizedSearchCV, CV=5) ì‹œìž‘ ---")
start_time = time.time()

# 1. ê³¼ì í•© ë°©ì§€ë¥¼ ìœ„í•œ íƒìƒ‰ ê³µê°„ ì •ì˜ (íŒŒë¼ë¯¸í„° ë²”ìœ„ ì¡°ì •)
param_dist_optimized = {
    # ë³µìž¡ë„ ì œì–´: max_depth ìƒí•œì„ 7ë¡œ ë‚®ì¶¤
    'max_depth': randint(3, 8),
    # ë³µìž¡ë„ ì œì–´: min_child_weight í•˜í•œì„ ë†’ì—¬ íŠ¸ë¦¬ê°€ ì‰½ê²Œ ë¶„í• ë˜ì§€ ì•Šë„ë¡ í•¨
    'min_child_weight': randint(3, 15),
    # ë¶€ìŠ¤íŒ… ë¼ìš´ë“œ ìˆ˜ëŠ” 500 ë‚´ì—ì„œ ì ì ˆížˆ í•™ìŠµí•˜ë„ë¡ ìœ ì§€
    'n_estimators': randint(200, 500),
    'learning_rate': uniform(0.01, 0.1), # í•™ìŠµë¥  ë²”ìœ„ë¥¼ ì¢í˜€ ì•ˆì •í™” ìœ ë„
    'subsample': uniform(0.7, 0.3),
    'colsample_bytree': uniform(0.7, 0.3),
    'gamma': uniform(0, 0.3),           # gamma ë²”ìœ„ ì¶•ì†Œ

    # ðŸ’¡ í•µì‹¬: ì •ê·œí™” íŒŒë¼ë¯¸í„° ì¶”ê°€ (ê³¼ì í•© ë°©ì§€ ëª©ì )
    'lambda': uniform(0.5, 2),  # L2 ì •ê·œí™” (ë³´í†µ 1ì´ ê¸°ë³¸ê°’)
    'alpha': uniform(0, 0.5)    # L1 ì •ê·œí™”
}

# 2. XGBoost ëª¨ë¸ ì´ˆê¸°í™” (ì •ê·œí™” íŒŒë¼ë¯¸í„°ëŠ” íŠœë‹ ëŒ€ìƒì— í¬í•¨)
# ë² ì´ìŠ¤ë¼ì¸ xgb_model ì¸ìŠ¤í„´ìŠ¤ ì‚¬ìš©
xgb_model.set_params(eval_metric='logloss', random_state=42, n_jobs=-1)

# 3. Stratified K-Fold ì •ì˜ (í´ëž˜ìŠ¤ ë¶ˆê· í˜• ê³ ë ¤)
# CV=5ë¡œ ì„¤ì •í•˜ì—¬ ê²€ì¦ ì‹ ë¢°ë„ í–¥ìƒ
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 4. RandomizedSearchCV ì„¤ì • ë° ì‹¤í–‰
random_search_optimized = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist_optimized,
    n_iter=50, # ì´ 50íšŒ ì‹œë„ (50 x 5-fold = 250íšŒ í•™ìŠµ)
    scoring='roc_auc',
    cv=cv,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

random_search_optimized.fit(X_train_processed, y_train)

end_time = time.time()
print(f"\níŠœë‹ ì´ ì†Œìš” ì‹œê°„: {end_time - start_time:.2f} ì´ˆ")

# 5. ê²°ê³¼ ì¶œë ¥
print("\n--- ê³¼ì í•© ë°©ì§€ íŠœë‹ ê²°ê³¼ ---")
print(f"ìµœì  CV AUC-ROC ì ìˆ˜: {random_search_optimized.best_score_:.4f}")
print("ìµœì  í•˜ì´í¼íŒŒë¼ë¯¸í„° ì¡°í•©:")
print(random_search_optimized.best_params_)

best_xgb_model_optimized = random_search_optimized.best_estimator_
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
import time

# X_train_processed, y_trainì€ ì´ë¯¸ ì „ì²˜ë¦¬ ì™„ë£Œëœ ë°ì´í„°ë¼ê³  ê°€ì •í•©ë‹ˆë‹¤.

# 1. ì´ì „ ìµœì í™”ëœ XGBoost íŒŒë¼ë¯¸í„°ë¥¼ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ ì •ì˜
# GridSearchCVë¥¼ í†µí•´ ì–»ì€ ìµœì¢… ìµœì  íŒŒë¼ë¯¸í„° ì‚¬ìš©
final_xgb_params = {'alpha': 0.2, 'lambda': 1.0, 'learning_rate': 0.05,
                    'max_depth': 7, 'n_estimators': 550,
                    'min_child_weight': 14, 'subsample': 0.83,
                    'colsample_bytree': 0.70, 'gamma': 0.0,
                    'eval_metric': 'logloss', 'random_state': 42, 'n_jobs': -1}

xgb_final = XGBClassifier(**final_xgb_params)

# 2. LightGBM ëª¨ë¸ ì •ì˜ (XGBoostì™€ í•™ìŠµ ë°©ì‹ì´ ë‹¤ë¦„)
lgbm = LGBMClassifier(n_estimators=500, learning_rate=0.05, random_state=42, n_jobs=-1)

# 3. Logistic Regression ëª¨ë¸ ì •ì˜ (ì„ í˜•ì„± ë³´ê°•)
log_reg = LogisticRegression(solver='liblinear', random_state=42)

# ê¸°ë³¸ ëª¨ë¸ ë¦¬ìŠ¤íŠ¸ ì •ì˜ (ì´ ëª¨ë¸ë“¤ì˜ ì˜ˆì¸¡ ê²°ê³¼ê°€ Meta-Featureê°€ ë©ë‹ˆë‹¤.)
base_estimators = [
    ('xgb', xgb_final),
    ('lgbm', lgbm),
    ('lr', log_reg)
]

# êµì°¨ ê²€ì¦ ê°ì²´ ì •ì˜
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("--- ðŸ§  ìŠ¤íƒœí‚¹ ì•™ìƒë¸” ëª¨ë¸ ì •ì˜ ì™„ë£Œ ---")

# ìµœì¢… ì˜ˆì¸¡ ëª¨ë¸ (Meta Learner) ì •ì˜
final_estimator = LogisticRegression(solver='liblinear', random_state=42)

# StackingClassifier ì •ì˜
stacking_model = StackingClassifier(
    estimators=base_estimators,
    final_estimator=final_estimator,
    cv=cv,
    stack_method='predict_proba',
    passthrough=False,
    n_jobs=-1
)

print("\n--- ðŸš€ Stacking Ensemble í•™ìŠµ ì‹œìž‘ ---")
start_time = time.time()

# Stacking ëª¨ë¸ í•™ìŠµ (ëª¨ë“  ê¸°ë³¸ ëª¨ë¸ ë° ìµœì¢… ëª¨ë¸ í•™ìŠµ í¬í•¨)
# X_train_processedì™€ y_trainì´ ì‚¬ìš© ê°€ëŠ¥í•˜ë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤.
stacking_model.fit(X_train_processed, y_train)

end_time = time.time()
print(f"âœ… Stacking Ensemble ì´ ì†Œìš” ì‹œê°„: {end_time - start_time:.2f} ì´ˆ")

# 3. Stacking ëª¨ë¸ ì„±ëŠ¥ í‰ê°€ (Train AUC í™•ì¸)
y_pred_stacking = stacking_model.predict_proba(X_train_processed)[:, 1]
auc_stacking = roc_auc_score(y_train, y_pred_stacking)

print(f"\n--- ðŸŒŸ Stacking Ensemble ê²°ê³¼ ---")
print(f"Stacking Ensemble (Train) AUC-ROC: {auc_stacking:.4f}")
```
