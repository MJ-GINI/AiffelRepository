# 전처리

# 피처 생성

```python
# 나이 구간화 (Binning)
# 나이에 따른 비선형적인 관계를 모델이 학습

bins = [10*i for i in range(2, 10)]
labels = [f'{i}-{i+9}' for i in bins[:-1]]

train['age_group'] = pd.cut(train['age'], bins=bins, labels=labels, right=False)
test['age_group'] = pd.cut(test['age'], bins=bins, labels=labels, right=False)

print("Train DataFrame 'age_group' head:")
print(train['age_group'].head())
print("\nTest DataFrame 'age_group' head:")
print(test['age_group'].head())
```

```python
# 체형 관련 파생 변수 생성

train['BMI'] = train['weight(kg)'] / ((train['height(cm)'] / 100) ** 2)
test['BMI'] = test['weight(kg)'] / ((test['height(cm)'] / 100) ** 2)

train['waist_to_height_ratio'] = train['waist(cm)'] / train['height(cm)'] # 추가
test['waist_to_height_ratio'] = test['waist(cm)'] / test['height(cm)']

train['weight_waist_interaction'] = train['weight(kg)'] * train['waist(cm)']
test['weight_waist_interaction'] = test['weight(kg)'] * test['waist(cm)']

train['BMI_AST_interaction'] = train['BMI'] * train['AST']
test['BMI_AST_interaction'] = test['BMI'] * test['AST']

train['BMI_ALT_interaction'] = train['BMI'] * train['ALT']
test['BMI_ALT_interaction'] = test['BMI'] * test['ALT']

train['BMI_Gtp_interaction'] = train['BMI'] * train['Gtp']
test['BMI_Gtp_interaction'] = test['BMI'] * test['Gtp']
```

```python
# 콜레스테롤 관련 파생 변수 생성

train['HDL_risk'] = train['HDL'] / train['LDL'] # 높을수록 좋다(???)
test['HDL_risk'] = test['HDL'] / test['LDL']

train['TG_risk'] = train['triglyceride'] / train['HDL'] # 낮을수록 좋다(???)
test['TG_risk'] = test['triglyceride'] / test['HDL']
```

```python
# 시력 및 청력 관련 파 변수 생성

train['avg_eyesight'] = (train['eyesight(left)'] + train['eyesight(right)']) / 2
test['avg_eyesight'] = (test['eyesight(left)'] + test['eyesight(right)']) / 2

train['impaired_eyesight'] = (train['avg_eyesight'] < 0.5).astype(int)
test['impaired_eyesight'] = (test['avg_eyesight'] < 0.5).astype(int)

train['impaired_hearing'] = ((train['hearing(left)'] != 1) | (train['hearing(right)'] != 1)).astype(int)
test['impaired_hearing'] = ((test['hearing(left)'] != 1) | (test['hearing(right)'] != 1)).astype(int)
```

```python
# 혈압 관련 파생 변수 생성

train['pulse_pressure'] = train['systolic'] - train['relaxation']
test['pulse_pressure'] = test['systolic'] - test['relaxation']
```

```python
# 로그 변환 'Gtp'

import numpy as np

# 'Gtp' 칼럼에 로그 변환 적용
train['Gtp'] = np.log1p(train['Gtp'])
test['Gtp'] = np.log1p(test['Gtp'])

print("Train DataFrame 'Gtp' after log transformation (head):")
display(train['Gtp'].head())
print("\nTest DataFrame 'Gtp' after log transformation (head):")
display(test['Gtp'].head())
```

# 데이터 스케일링 및 인코딩

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

# 타겟 변수 분리
X_train = train.drop('smoking', axis=1)
y_train = train['smoking']
X_test = test.copy()

# 숫자형 및 범주형 피처 정의
numerical_features = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = ['age_group']

# ColumnTransformer 설정
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# 전처리 파이프라인 생성 및 적용
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# OneHotEncoder로 생성된 새로운 피처 이름 가져오기
encoded_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)

# 모든 피처 이름 조합
all_feature_names = numerical_features + encoded_feature_names.tolist()

# 처리된 데이터를 DataFrame으로 변환 (선택 사항, 모델에 따라 numpy 배열로 사용 가능)
X_train_df = pd.DataFrame(X_train_processed, columns=all_feature_names)
X_test_df = pd.DataFrame(X_test_processed, columns=all_feature_names)

print("X_train_df head after scaling and encoding:")
display(X_train_df.head())
print("\nX_test_df head after scaling and encoding:")
display(X_test_df.head())
```

```python
print("train DataFrame 인코딩 이후")
print(train.head())
print("train DataFrame 인코딩 이후")
train.info()

print("test DataFrame 인코딩 이후")
print(test.head())
print("test DataFrame 인코딩 이후")
test.info()
```

# 모델링

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

# lr
log_reg_model = LogisticRegression(solver='liblinear', random_state=42, n_jobs=-1)

# rf
rf_model = RandomForestClassifier(random_state=42, n_jobs=-1)

# Igbm
lgbm_model = LGBMClassifier(random_state=42, n_jobs=-1)

# xgb
xgb_model = XGBClassifier(random_state=42, n_jobs=-1, eval_metric='logloss')

# Train lr
print("Training Logistic Regression model...")
log_reg_model.fit(X_train_processed, y_train)
print("Logistic Regression model trained.")

# Train rf
print("Training RandomForestClassifier model...")
rf_model.fit(X_train_processed, y_train)
print("RandomForestClassifier model trained.")

# Train lgbm
print("Training LGBMClassifier model...")
lgbm_model.fit(X_train_processed, y_train)
print("LGBMClassifier model trained.")

# Train xgb
print("Training XGBClassifier model...")
xgb_model.fit(X_train_processed, y_train)
print("XGBClassifier model trained.")
```

```python
from sklearn.metrics import roc_auc_score

# Evaluate lr model
y_pred_log_reg = log_reg_model.predict_proba(X_train_processed)[:, 1]
auc_log_reg = roc_auc_score(y_train, y_pred_log_reg)
print(f"Logistic Regression AUC-ROC: {auc_log_reg:.4f}")

# Evaluate rf model
y_pred_rf = rf_model.predict_proba(X_train_processed)[:, 1]
auc_rf = roc_auc_score(y_train, y_pred_rf)
print(f"RandomForestClassifier AUC-ROC: {auc_rf:.4f}")

# Evaluate lgbm model
y_pred_lgbm = lgbm_model.predict_proba(X_train_processed)[:, 1]
auc_lgbm = roc_auc_score(y_train, y_pred_lgbm)
print(f"LGBMClassifier AUC-ROC: {auc_lgbm:.4f}")

# Evaluate xgb model
y_pred_xgb = xgb_model.predict_proba(X_train_processed)[:, 1]
auc_xgb = roc_auc_score(y_train, y_pred_xgb)
print(f"XGBClassifier AUC-ROC: {auc_xgb:.4f}")
```

## 이미 실행해서 재실행하지 않은 코드

### 랜덤그리드서치

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint

# XGBoost 모델의 하이퍼파라미터 탐색 공간 정의
# 이전 훈련에서 사용한 XGBClassifier 인스턴스를 직접 사용

param_dist = {
    'n_estimators': randint(100, 500),  # 부스팅 라운드 수
    'learning_rate': uniform(0.01, 0.2), # 학습률
    'max_depth': randint(3, 10),        # 트리의 최대 깊이
    'subsample': uniform(0.6, 0.4),     # 샘플 비율
    'colsample_bytree': uniform(0.6, 0.4), # 피처 샘플 비율
    'gamma': uniform(0, 0.5),           # 리프 노드 분할에 필요한 최소 손실 감소
    'min_child_weight': randint(1, 10) # 자식 노드에 필요한 최소 가중치 합
}

# RandomizedSearchCV 설정
# cv=3은 빠르게 탐색하기 위함 (실제로는 5 이상 권장)
# n_iter=50은 탐색할 하이퍼파라미터 조합의 수 (빠른 확인을 위함)
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=50,
    scoring='roc_auc',
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=-1 # 모든 코어 사용
)

print("Starting RandomizedSearchCV for XGBoost model...")
# 튜닝 실행
random_search.fit(X_train_processed, y_train)

print("RandomizedSearchCV completed.")
print(f"Best AUC-ROC score: {random_search.best_score_:.4f}")
print("Best parameters found:")
display(random_search.best_params_)
```

### 그리드서치

## 최적 파라미터로 XGBoost 모델 재학습 및 평가

```python
from sklearn.metrics import roc_auc_score
import xgboost as xgb

# 최적 파라미터로 XGBoost 모델 초기화 (각 파라미터를 명시적으로 설정)
best_xgb_model_tuned = xgb.XGBClassifier(
    random_state=42,
    n_jobs=-1,
    eval_metric='logloss',
    n_estimators=375,
    learning_rate=0.05,
    max_depth=9,
    subsample=0.8,
    colsample_bytree=0.75,
    gamma=0.3,
    min_child_weight=1
)

print("\nTraining XGBoost model with best parameters...")
# 모델 학습
best_xgb_model_tuned.fit(X_train_processed, y_train)
print("XGBoost model trained with best parameters.")

# 훈련 데이터에 대한 예측 확률 및 AUC-ROC 계산
y_pred_best_xgb_train_tuned = best_xgb_model_tuned.predict_proba(X_train_processed)[:, 1]
auc_best_xgb_train_tuned = roc_auc_score(y_train, y_pred_best_xgb_train_tuned)
print(f"\nTuned XGBoost (Train) AUC-ROC: {auc_best_xgb_train_tuned:.4f}")

# 테스트 데이터에 대한 예측 확률 생성
y_pred_best_xgb_test_tuned = best_xgb_model_tuned.predict_proba(X_test_processed)[:, 1]

print("\nTuned XGBoost (Test) predictions head:")
print(y_pred_best_xgb_test_tuned[:5])
```

# 성능이 좋아도 문제 ;; 과적합인지 확인

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import roc_auc_score

# 과적합 여부를 확인하기 위해
# 훈련 데이터!!!!!!!!!!를 훈련 세트와 검증 세트로 분리
# test_size는 검증 세트의 비율 (예: 20%)
# stratify=y_train은 타겟 변수의 클래스 비율을 유지하면서 분할합니다.
# random_state는 재현성을 위해 고정합니다.
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train_processed, y_train, test_size=0.2, random_state=42, stratify=y_train
)

print(f"Original training data shape: {X_train_processed.shape}")
print(f"New training data shape: {X_train_split.shape}")
print(f"Validation data shape: {X_val_split.shape}")

# 최적 파라미터로 XGBoost 모델 초기화 (이전 튜닝에서 찾은 최적 파라미터 사용)
best_xgb_model_tuned_for_validation = xgb.XGBClassifier(
    random_state=42,
    n_jobs=-1,
    eval_metric='logloss',
    n_estimators=375,
    learning_rate=0.05,
    max_depth=9,
    subsample=0.8,
    colsample_bytree=0.75,
    gamma=0.3,
    min_child_weight=1
)

print("\nTraining XGBoost model on new training split...")
# 분리된 훈련 세트로 모델 학습
best_xgb_model_tuned_for_validation.fit(X_train_split, y_train_split)
print("XGBoost model trained on new training split.")

# 새로운 훈련 세트에 대한 예측 확률 및 AUC-ROC 계산
y_pred_train_split = best_xgb_model_tuned_for_validation.predict_proba(X_train_split)[:, 1]
auc_train_split = roc_auc_score(y_train_split, y_pred_train_split)
print(f"\n(New Split) Tuned XGBoost (Train) AUC-ROC: {auc_train_split:.4f}")

# 검증 세트에 대한 예측 확률 및 AUC-ROC 계산
y_pred_val_split = best_xgb_model_tuned_for_validation.predict_proba(X_val_split)[:, 1]
auc_val_split = roc_auc_score(y_val_split, y_pred_val_split)
print(f"(New Split) Tuned XGBoost (Validation) AUC-ROC: {auc_val_split:.4f}")

# 훈련 세트와 검증 세트의 AUC-ROC 점수를 비교하여 과적합 여부를 판단
# 두 점수 차이가 크지 않다면 모델이 일반화 성능이 양호
# 근데 0.1차이? 큰 것임...
```
