## 연습문제1. 환율 계산기
- 이 계산기는 사용자로부터 원화 금액을 입력받아 지정된 환율로 미국 달러(USD)로 변환하는 역할을 합니다.

- 요구사항:
1. 환율 정의 (Lambda):
    - 현재 환율 1USD = 1,350 KRW를 사용하여, 원화(KRW)를 달러(USD)로 변환하는 람다 함수를 정의하세요.
    - krw_to_usd = lambda amount: ... 형태로 정의해야 합니다.
2. 함수 정의 및 예외 처리 (Function & Exception):
    - calculate_exchange(amount)라는 일반 함수를 정의하세요.
    - 이 함수는 변환할 원화 금액(amount)을 입력받습니다.
    - 금액(amount)이 음수이거나 0일 경우, ValueError를 발생시키고 "입력 금액은 0보다 커야 합니다."라는 메시지를 출력하도록 예외 처리를 구현하세요.
    - 정상적인 금액일 경우, 위에서 정의한 람다 함수를 사용하여 변환 결과를 반환합니다.
3. 호출 및 테스트:
    - 함수를 호출하여 270,000 KRW를 변환해보고, 예외 상황(-100 KRW)도 테스트해보세요.

```python
# 1. 환율 정의
# 원화(KRW)를 달러(USD)로 변환하는 람다 표현식 정의
exchange_rate = 1350
krw_to_usd = lambda krw_amount : krw_amount / exchange_rate

# 2. 함수 및 예외 처리 정의
# calculate_exchange(amount)라는 일반 함수 정의
def calculate_exchange(amount):
    # 금액(amount)이 음수이거나 0일 경우, ValueError를 발생시키고
    # "입력 금액은 0보다 커야 합니다."라는 메시지를 출력하도록 예외 처리를 구현
    if amount <= 0:
        raise ValueError("입력 금액은 0보다 커야 합니다.")
    
    # 람다 표현식 사용하여 환전
    usd_amount = krw_to_usd(amount)
    
    # 결과를 소수점 2자리까지 표시
    return round(usd_amount, 2)

# 테스트
# 3-1. 정상적인 금액 27만원일 때
try: 
    krw_input = 270000
    usd_result = calculate_exchange(krw_input)
    print(f"{krw_input:,.0f} KRW는 {usd_result:,.2f} USD 입니다.")
except ValueError as e:
    print(f"오류 발생: {e}")

# 3-2. 예외 상황 -100원일 때
try:
    krw_input_error = -100
    usd_result_error = calculate_exchange(krw_input_error)
    print(f"{krw_input_error:,.0f} KRW는 {usd_result_error:,.2f} USD 입니다.")
except ValueError as e:
    print(f"오류 발생: {e}")

# 3-3. 0원 일때
try:
    krw_input_zero = 0
    usd_result_zero = calculate_exchange(krw_input_zero)
    print(f"{krw_input_zero:,.0f} KRW는 {usd_result_zero:,.2f} USD 입니다.")
except ValueError as e:
    print(f"오류 발생: {e}")
```

## 연습문제2. 차량 관리 시스템 및 정렬

당신은 회사 차량을 관리하는 간단한 시스템을 개발해야 합니다. 
모든 차량은 기본적인 정보를 가지고 있지만, 특정 유형의 차량(예: 전기차)은 추가적인 속성을 가집니다.

- 요구 사항
1. 기본 클래스 정의:
    - Vehicle이라는 부모 클래스를 정의하세요.
    - 속성(attribute): model(모델명), year(제조년도)
    - 메서드(method): display_info() 차량의 기본 정보를 출력합니다.
2. 상속 클래스 정의(Module/Class)
    - Vehicle 클래스를 상속받는 ElectricVehicle이라는 자식 클래스를 정의하세요.
    - 추가 속성: battery_range (배터리 주행 가능 거리, km 단위)
    - display_info() 메서드를 오버라이딩하여, 기본 정보 외에 배터리 주행 거리도 함께 출력하도록 만드세요.
3. 람다 함수를 이용한 정렬
    - 몇 대의 ElectricVehicle 객체를 생성하여 리스트에 저장하세요.
    - 이 리스트를 람다 함수를 사용하여 battery_range 속성 기준으로 내림차순 정렬한 후 결과를 출력하세요.

```python
# 1. 기본(부모) 클래스 정의
class Vehicle:
    # 차량 기본 정보 속성
    def __init__ (self, model, year):
        self.model = model
        self.year = year
    
    # 차량 기본 정보 메서드
    def display_info(self):
        print(f"모델: {self.model}, 제조년도: {self.year}")

# 2. 상속(자식) 클래스 정의
class ElectricVehicle(Vehicle):
    def __init__(self, model, year, battery_range):
        # 부모 클래스 생성자 호출
        super().__init__(model, year)
        # 추가 속성
        self.battery_range = battery_range
        
    # display_info() 메서드 덮어쓰기(오버라이딩)
    def display_info(self):
        super().display_info()
        print(f"배터리 주행 가능 거리: {self.battery_range} km")
        print("-"*30)
        
# 3. ElectricVehicle 객체 생성
cars = [
    ElectricVehicle("테슬라 모델3", 2025, 513),
    ElectricVehicle("기아 EV9", 2025, 501),
    ElectricVehicle("폴스타2 싱글모터", 2025, 511),
    ElectricVehicle("아이오닉 6N", 2025, 614)

]

print("1. 전기차 기본 정보 목록\n")
for car in cars:
    car.display_info()
```

```python
# 4. 람다 함수를 이용한 정렬
# 각 객체에서 battery_range 속성 값만 반환 -> 일급 객체
sort_lambda = lambda car_object: car_object.battery_range

# 내림차순 정렬
# sorted(리스트): 새로운 정렬된 리스트를 반환하는 python 내장 함수
# key= : 정렬 기준으로 사용할 값을 추출하는 함수를 지정
# reverse=True: Descending 기준 정렬
sorted_cars = sorted(cars, key=sort_lambda, reverse=True)

print("2. 배터리 주행 가능 거리 기준 정렬(내림차순)\n")
for car in sorted_cars:
    car.display_info()
```
