### 027. 10진 정수 입력받아 16진수로 출력하기1


```python
#10진수를 입력받아 16진수(hexadecimal)로 출력해보자.

#예시
a = input()
n = int(a)      #입력된 a를 10진수 값으로 변환해 변수 n에 저장
print('%x'% n)  #n에 저장되어있는 값을 16진수(hexadecimal) 소문자 형태 문자열로 출력
```

```python
# 10진수 입력
a = input()
n = int(a)

# 16진수 출력
print(hex(n))
```

| 구분     | 10진수     | 16진수                        |
| :------ | :-------- | :--------------------------- |
| 기호     | 0~9      | 0~9, A~F                    |
| 자리값    | 10의 거듭제곱 | 16의 거듭제곱                    |
| 사용처    | 일상 생활    | 컴퓨터 메모리, 색상코드, 하드웨어         |
| 파이썬 표현 | 255      | 0xFF, format(n, 'x') → 'ff' |

* 10진수는 인간 친화적
* 16진수는 컴퓨터 내부 데이터 표현에 많이 사용됨

```python
# 태어난 해, 월, 일을 입력받아 사주팔자를 보는 프로그램을 작성하시오.

# 사주를 보는 방법:

# 세 수(년,월,일)가 주어지면,
# (년 - 월 + 일)에 마지막 숫자가 2이면 "대박"을 출력, 그렇지 않으면 "그럭저럭"을 출력하세요.
```

### 1164. 터널 통과하기

```python
# 어떤 차의 높이가 170cm 이다.

# 이 차는 3개의 터널을 차례대로 지나게 될 것이다.

# 터널의 높이가 차의 높이보다 같거나 낮다면 차는 터널과 충돌하여 사고가 날 것이다.

# 터널의 높이가 차례대로 3개 주어지면 터널을 무사히 잘 통과하면 PASS 를 출력하고, 사고가 난다면 CRASH 를 출력하시오.
```

```python
#차의 높이
car_height = 170

#터널 높이 (공백 구분)
t1, t2, t3 = map(int, input().split())

#터널별 높이를 리스트로 변수 생성
tunnels = [t1, t2, t3]

#for index, value in enumerate(iterable, start=0)
#index: 현재 요소의 위치
#value: 현재 요소의 값
#enumerate: 반복 가능한 객체를 반복할 때 인덱스와 값을 동시에 가져오는 함수
#iterable: 반복 가능한 객체
#start=n: 인덱스 시작값 지정 = 인덱스를 n부터 시작
#i = 반복횟수 또는 터널번호 (1부터 시작)
#t = 해당 반복에서의 터널 높이 값

for i, t in enumerate(tunnels, start=1):
    if car_height >= t:
        print(f"터널 {i}: CRASH")
    else:
        print(f"터널 {i}: PASS")
```

### 1226. 이번 주 로또

지혜는 로또 매니아다.

매주 로또 한장을 사고 토요일이면 대박을 기대하면서 당첨번호를 확인한다.

지혜는 로또 결과가 4등이었는데 확인을 잘못해서 5등인줄 알고 그냥 바꾸기 귀찮아서 버렸었다.

이러한 사태가 다시는 일어나지 않게 하기 위해 우리가 프로그램을 만들어 주자.

```python
#1. 당첨번호 6개와 보너스 번호 입력
#당첨번호 입력 > 공백 분리 > 정수 변환 map(int()) > list로 묶어주기
winning_numbers = list(map(int, input("이번 주 당첨번호: ").split()))

#당첨번호 7개 중 1~6번까지만 슬라이싱해서 변수(winning) 저장
winning = winning_numbers[:6]

#당첨번호 7개 중 마지막 7번 슬라이싱해서 변수(bonus) 저장
bonus = winning_numbers[6] 


#2. 나의 로또 번호 입력
my_nums = list(map(int, input("당신의 로또번호: ").split()))


#3. 당첨번호와 일치하는 개수 확인
#당첨번호(winning)와 나의 번호(my_nums)를 각각의 집합(set)으로 바꾸기
#집합(set): 순서 없음, 중복 불가
#[3 13 15 24 33 37 2] -> {3 13 15 24 33 37 2}

#교집합(&) > len() 교집합의 요소 개수를 셈

#변수(match_cnt)저장
match_cnt = len(set(winning) & set(my_nums))

#4. 보너스 번호 확인
#bonus 변수에 my_nums에 있는(in) 숫자가 있는지 확인 > True/False Bool 반환 > 변수(has_sbonus) 저장
has_bonus = bonus in my_nums

#5. 로또 결과 등수 판정
if match_cnt == 6: 
    #1등 : 6개 모두 일치
    print("1등")
elif match_cnt == 5 and has_bonus: 
    #2등: 5개 일치 + 보너스 번호
    print("2등")
elif match_cnt == 5: 
    #3등: 5개 일치
    print("3등")
elif match_cnt == 4: 
    #4등: 4개 일치
    print("4등")
elif match_cnt == 3: 
    #5등: 3개 일치
    print("5등")
else: 
    #꽝: 그 외
    print("꽝")
```
